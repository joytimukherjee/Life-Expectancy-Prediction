```python
from sklearn.impute import KNNImputer
impute=KNNImputer()
```

```python
for i in df.select_dtypes(include="number").columns:
    df[i]=impute.fit_transform(df[[i]])
```

```python
#outlier treatment
def wisker (col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return q1,q3
```

```python
wisker(df['GDP'])
```

```python
df.columns
```

```python
for i in df.select_dtypes(include="number").columns:
    lw, uw = wisker(df[i])
    df[i] = np.where(df[i] < lw, lw, df[i])
    df[i] = np.where(df[i] > uw, uw, df[i])
```

```python
#encoding of data
df = pd.get_dummies(data=df, columns=["Status"], drop_first=True)
df.drop(columns=['Country'], inplace=True)
```

```python
df.columns[df.columns.str.startswith("Status_")]
```

```python
df.columns = df.columns.str.strip()  # remove extra spaces
df.columns = df.columns.str.replace(' +', ' ', regex=True)  # fix double spaces
```

```python
# feature engineering

features=[ 'Year',  'Adult Mortality',
       'infant deaths', 'Alcohol', 'percentage expenditure', 'Hepatitis B',
       'Measles', 'BMI', 'Polio', 'Total expenditure',
        'HIV/AIDS', 'GDP', 'Population', 'thinness 1-19 years',
       'thinness 5-9 years', 'Income composition of resources', 'Schooling','Status_Developing']
```

```python
df_selected= df[features+["Life expectancy"]]
X=df_selected[features] #x
y=df_selected["Life expectancy"] #y needs prediction
```

```python
x=X.apply(pd.to_numeric,errors="coerce") #converts to numeric if any in str and if cannot convert tha with Nan values
X=X.fillna(0)
```

```python
#scaling
from sklearn.preprocessing import StandardScaler
scaler =StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# train  and split
from sklearn.model_selection import train_test_split
X_train ,X_test ,y_train,y_test = train_test_split(X_scaled,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor  
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Define models
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42)
}

# Example usage:
# Assuming you have your data (X_train, y_train, X_test, y_test)
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Calculate metrics
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    
    print(f"Model: {name}")
    print(f"Mean Absolute Error: {mae}")
    print(f"Mean Squared Error: {mse}")
    print(f"Root Mean Squared Error: {rmse}")
    print(f"R-squared (Accuracy): {r2}")


```

```python
#model train
best_model=RandomForestRegressor(n_estimators=100,random_state=42)
best_model.fit(X_train,y_train)
y_pred=best_model.predict(X_test)
```

```python
#model evaluation
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_test,y= y_pred)
plt.xlabel('Actual Life Expectancy')
plt.ylabel('Predicted Life Expectancy')
plt.title('Actual vs Predicted Life Expectancy')
plt.show()
```

```python
#feature Importance Visualization
importances=best_model.feature_importances_
feature_names=features

plt.figure(figsize=(8, 6))
sns.barplot(x=importances, y=feature_names)
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.title("Feature Importances in Prediction of Life Expectancy")
plt.show()
```
